```python
#%%script echo skipping

# @title Criteria based on the daily Tmin values
import numpy as np
import pandas as pd
from scipy.stats import linregress
def mean_squared_log_error(Si, Oi):
    non_zero_mask = (Si != 0) & (Oi != 0)
    Si_non_zero = Si[non_zero_mask]
    Oi_non_zero = Oi[non_zero_mask]

    msle = np.mean(np.power(np.log(Si_non_zero) - np.log(Oi_non_zero), 2))
    return msle

# Define a list of climate models and scenarios
climate_models = ["Climate_model_1", "Climate_model_2", "Climate_model_3", "Climate_model_4", "Climate_model_5",
                          "Climate_model_6", "Climate_model_7", "Climate_model_8", "Climate_model_9", "Climate_model_10",
                          "Climate_model_11"]
scenarios = ['rcp45']
# Create an empty DataFrame to store the data
Tmin_df = pd.DataFrame()
for clmodel in climate_models:
    for scenario in scenarios:


      meteo_directory = '/content/drive/MyDrive/Comparisons/Meteo'
      os.makedirs(meteo_directory, exist_ok=True)

      eurocordex_directory = '/content/drive/MyDrive/Comparisons/EuroCordex'
      os.makedirs(eurocordex_directory, exist_ok=True)
      eurocordex_combined_file = "/content/drive/MyDrive/Comparisons/EuroCordex/Daily/"+clmodel+"/EuroCordex_"+scenario+"_combined.csv"
      meteo_combined_file = "/content/drive/MyDrive/Comparisons/Meteo/Daily/Meteo_combined.csv"


      # Load CSV data into DataFrames
      df_eurocordex = pd.read_csv(eurocordex_combined_file)
      df_meteo = pd.read_csv(meteo_combined_file)

      # Identify common rows based on 'Month', 'Year', and 'Location'
      common_rows = pd.merge(df_eurocordex[['Month', 'Year', 'Location']], df_meteo[['Month', 'Year', 'Location']], how='inner', on=['Month', 'Year', 'Location'])

      # Identify rows with '-' in either dataset
      rows_with_dash_eurocordex = df_eurocordex[df_eurocordex.apply(lambda row: '-' in row.values, axis=1)]
      rows_with_dash_meteo = df_meteo[df_meteo.apply(lambda row: '-' in row.values, axis=1)]

      # Identify common rows with '-'
      common_rows_with_dash = pd.merge(rows_with_dash_eurocordex[['Month', 'Year', 'Location']], rows_with_dash_meteo[['Month', 'Year', 'Location']], how='inner', on=['Month', 'Year', 'Location'])

      # Remove common rows with '-'
      df_eurocordex = df_eurocordex[~df_eurocordex.index.isin(common_rows_with_dash.index)]
      df_meteo = df_meteo[~df_meteo.index.isin(common_rows_with_dash.index)]

      df_eurocordex['Tmin'] = pd.to_numeric(df_eurocordex['Tmin'], errors='coerce')
      df_meteo['Tmin'] = pd.to_numeric(df_meteo['Tmin'], errors='coerce')


      # Filter data to keep only rows with Month values 1, 2, 3, 4, 5, 6, 7, and 12
      months_to_keep = [1, 2, 3, 4, 5, 6, 7, 12]
      df_eurocordex = df_eurocordex[df_eurocordex['Month'].isin(months_to_keep)]
      df_eurocordex = df_eurocordex[~((df_eurocordex['Year'] == 2023) & (df_eurocordex['Month'].isin([8,9,10,11,12])))]

      # Conditions for specific Locations in df_eurocordex
      conditions_eurocordex = {
          'Kopaida': ((df_eurocordex['Location'] == 'Kopaida') & (df_eurocordex['Year'] == 2014) & (df_eurocordex['Month'] < 12)),
          'Livadeia': ((df_eurocordex['Location'] == 'Livadeia') & (df_eurocordex['Year'] == 2009) & (df_eurocordex['Month'] < 12)),
          'Lefkochori': ((df_eurocordex['Location'] == 'Lefkochori') & (df_eurocordex['Year'] == 2009) & (df_eurocordex['Month'] < 12)),
          'Amfikleia_rural': ((df_eurocordex['Location'] == 'Amfikleia_rural') & (df_eurocordex['Year'] == 2006) & (df_eurocordex['Month'] < 12)),
          'Amfikleia_urban': ((df_eurocordex['Location'] == 'Amfikleia_urban') & (df_eurocordex['Year'] == 2008) & (df_eurocordex['Month'] < 12)),
          'Polydroso': ((df_eurocordex['Location'] == 'Polydroso') & (df_eurocordex['Year'] == 2010) & (df_eurocordex['Month'] < 12)),
          'Antikyra': ((df_eurocordex['Location'] == 'Antikyra') & (df_eurocordex['Year'] == 2013) & (df_eurocordex['Month'] < 12)),
          'Kyriaki': ((df_eurocordex['Location'] == 'Kyriaki') & (df_eurocordex['Year'] == 2016) & (df_eurocordex['Month'] < 12))
      }

      # Apply conditions and remove rows in df_eurocordex
      for location, condition in conditions_eurocordex.items():
          df_eurocordex = df_eurocordex[~condition]



      # Filter data to keep only rows with Month values 1, 2, 3, 4, 5, 6, 7, and 12
      df_meteo = df_meteo[df_meteo['Month'].isin(months_to_keep)]
      df_meteo = df_meteo[~((df_meteo['Year'] == 2023) & (df_meteo['Month'].isin([8,9,10,11,12])))]

      # Conditions for specific Locations in df_meteo
      conditions_meteo = {
          'Kopaida': ((df_meteo['Location'] == 'Kopaida') & (df_meteo['Year'] == 2014) & (df_meteo['Month'] < 12)),
          'Livadeia': ((df_meteo['Location'] == 'Livadeia') & (df_meteo['Year'] == 2009) & (df_meteo['Month'] < 12)),
          'Lefkochori': ((df_meteo['Location'] == 'Lefkochori') & (df_meteo['Year'] == 2009) & (df_meteo['Month'] < 12)),
          'Amfikleia_rural': ((df_meteo['Location'] == 'Amfikleia_rural') & (df_meteo['Year'] == 2006) & (df_meteo['Month'] < 12)),
          'Amfikleia_urban': ((df_meteo['Location'] == 'Amfikleia_urban') & (df_meteo['Year'] == 2008) & (df_meteo['Month'] < 12)),
          'Polydroso': ((df_meteo['Location'] == 'Polydroso') & (df_meteo['Year'] == 2010) & (df_meteo['Month'] < 12)),
          'Antikyra': ((df_meteo['Location'] == 'Antikyra') & (df_meteo['Year'] == 2013) & (df_meteo['Month'] < 12)),
          'Kyriaki': ((df_meteo['Location'] == 'Kyriaki') & (df_meteo['Year'] == 2016) & (df_meteo['Month'] < 12))
      }

      # Apply conditions and remove rows in df_meteo
      for location, condition in conditions_meteo.items():
          df_meteo = df_meteo[~condition]

          # Remove rows where 'Day' is 32 in df_meteo
      df_meteo = df_meteo[df_meteo['Day'] != 32]
      # Extract 'Tmin' and 'Location' values
      Tmin_eurocordex = df_eurocordex['Tmin']*0.0864
      Tmin_meteo = df_meteo['Tmin']



      # Add a column to the result_df with the EuroCordex data
      Tmin_df[f'Tmin_{clmodel}'] = Tmin_eurocordex


# Add the 'Tmin_meteo' column to the result_df

Tmin_df.insert(0, 'Tmin_meteo', Tmin_meteo)
# Remove rows with NaN in any column
Tmin_df = Tmin_df.dropna()
# Extracting column names from the dataframe
column_names = Tmin_df.columns

# Specify the column to be used for the y-axis
y_column = Tmin_df.columns[0]  # Assuming the 5th column is the 5th in the dataframe
# Define the margin
margin = 2

# Create an empty DataFrame to store the results
results_df = pd.DataFrame()

# Define O_i and P_i
Oi = Tmin_df[y_column].values
for i in range(1, 12):
    x_column = column_names[i]
    Si = Tmin_df[x_column].values
    # Check if the lengths of Oi and Si are the same
    min_len = min(len(Oi), len(Si))
    Oi = Oi[:min_len]
    Si = Si[:min_len]

     # Rest of the calculations remain the same
    AME = np.max(np.abs(Si - Oi))
    Peak_Difference = np.max(Si) - np.max(Oi)
    MAE = np.mean(np.abs(Si - Oi))
    ME = np.mean(Si - Oi)
    RMSE = np.sqrt(np.mean((Si - Oi)**2))
    R4MS4E = np.power(np.mean(np.power(np.abs(Si - Oi), 4)), 1/4)
    Sign_Changes = np.sum(np.diff(np.sign(Si - Oi)) != 0)
    RAE = np.sum(np.abs(Si - Oi)) / np.sum(np.abs(Oi - np.mean(Oi)))
    Percent_Error_in_Peak = np.abs(np.max(Si) - np.max(Oi)) / np.max(Oi) * 100
    MARE = np.mean(np.abs((Si - Oi) / Oi))#inf
    MdAPE = np.median(np.abs((Si - Oi) / Oi)* 100) #inf
    MRE = np.mean((Si - Oi) / Oi)#inf
    MSRE = np.mean(np.power((Si - Oi) / Oi, 2))#inf
    RVE = np.sum(Oi - Si) / np.sum(Oi)
    R_squared = np.sum((Oi - np.mean(Oi))*(Si - np.mean(Si))) / np.sqrt((np.sum(np.power(Oi - np.mean(Oi), 2)))*(np.sum(np.power(Si - np.mean(Si), 2))))
    Index_of_Agreement = 1 - np.sum((np.power(Si - Oi,2) )) / np.sum(np.power((np.abs(Oi - np.mean(Oi)) + np.abs(Si - np.mean(Oi))),2))
    NSE = 1 - np.sum(np.power(Oi - Si, 2)) / np.sum(np.power(Oi - np.mean(Oi), 2))
    Persistence_Index =  np.sum(np.power(Si - Oi, 2))/np.sum(np.power(Oi[1:] - Oi[:-1], 2))

    #MSLE = mean_squared_log_error(Si,Oi) #values>0 σε θερμοκρασίες εκτος

    # Modified line to handle lengths
    #MSDE = np.mean(np.power(Oi[1:] - 2*Oi[1:-1] + Oi[:-2] - Pi[1:] + 2*Pi[1:-1] - Pi[:-2], 2)[:min_len])

    #Inertia_RMSE = np.sqrt(np.mean(np.power(Oi[2:] - 2*Oi[1:-1] + Oi[:-2] - Si[2:] + 2*Si[1:-1] - Si[:-2], 2)))
    Volumetric_Efficiency = 1 - np.sum(np.abs(Si - Oi)) / np.sum(Oi)
    r = np.corrcoef(Si, Oi)[0, 1]
    a = np.std(Si) / np.std(Oi)
    b = np.mean(Si) / np.mean(Oi)
    KGE = 1 - np.sqrt((r-1)**2 + (a-1)**2 + (b-1)**2)
    SSE = np.sum(np.power(Si - Oi, 2))
    m = len(Oi)
    k = 2  # Assuming a simple linear regression model with intercept and slope
    AIC = m * np.log(RMSE) + 2 * k
    BIC = m * np.log(RMSE) + k * np.log(m)
    slope, intercept, _, _, _ = linregress(Si, Oi)
    R_squared_1to1 = slope**2

    # Append the results to the DataFrame
    results_df = results_df.append({
        'Criteria': x_column,
        'd': Index_of_Agreement,
        'NSE': NSE,
        'KGE': KGE,
        'VE': Volumetric_Efficiency,
        'AME': AME,
        'MAE': MAE,
        'MBE': ME,
        'RMSE': RMSE,
        'R4MS4E': R4MS4E,
        'RAE': RAE,
        'RVE': RVE,
        'SSE': SSE,
    }, ignore_index=True)


# Transpose the DataFrame
results_df_transposed = results_df.set_index('Criteria').transpose()
# Add a new column "Type" based on conditions
results_df_transposed['Type'] = np.where(
    results_df_transposed.index.isin(['d', 'NSE', 'KGE', 'VE']),
    'I',
    'II'
)
# Save the transposed DataFrame to an Excel file
results_df_transposed.to_excel(os.path.join(output_folder, 'Criteria_Tmin_Growing_season_'+scenario+'.xlsx'))


# Save the transposed DataFrame to an Excel file
results_df_transposed.to_csv(os.path.join(output_folder, 'Criteria_Tmin_Growing_season_'+scenario+'.csv'))
```

```python
#%%script echo skipping
# @title First transformation of daily values
import pandas as pd
import numpy as np

# Load the results DataFrame from the Excel file
results_df = pd.read_excel(os.path.join(output_folder, 'Criteria_Tmin_Growing_season_'+scenario+'.xlsx'))

# Create a new DataFrame for the transformed values
transformed_df = results_df.copy()

# Iterate over rows
for i in range(len(results_df)):
    row = results_df.iloc[i, 1:-1]  # Exclude the Type column
    row_type = results_df.loc[i, 'Type']

    if row_type == 'II':
        max_val = np.max(np.abs(row))
        transformed_row = 1 - (np.abs(row) + 1) / (np.abs(max_val) + 1)
        transformed_df.iloc[i, 1:-1] = transformed_row

# Save the transformed DataFrame to a new Excel file
transformed_df.to_excel(os.path.join(output_folder, 'First_Transformed_Criteria_Tmin_Growing_season_'+scenario+'.xlsx'), index=False)
```

```python
#%%script echo skipping
# @title Second transformation of daily values
import pandas as pd

# Load the original DataFrame from the Excel file
original_df = pd.read_excel(os.path.join(output_folder, 'First_Transformed_Criteria_Tmin_Growing_season_'+scenario+'.xlsx'))

# Create a new DataFrame to store the transformed values
transformed_df = original_df.copy()

# Iterate over each row
for i in range(len(original_df)):
    row_values = original_df.iloc[i, 1:-1]  # Exclude the 'Type' column
    min_val = np.min(row_values)
    max_val = np.max(row_values)

    # Apply the transformation to each value in the row
    transformed_values = (row_values - min_val) / (max_val - min_val)

    # Update the corresponding row in the new DataFrame
    transformed_df.iloc[i, 1:-1] = transformed_values

# Save the transformed DataFrame to a new Excel file
transformed_df.to_excel(os.path.join(output_folder, 'Second_Transformed_Criteria_Tmin_Growing_season_'+scenario+'.xlsx'), index=False)
```

```python
#%%script echo skipping
# @title RPI and sRPI
import pandas as pd
from scipy.stats import spearmanr
import numpy as np

# Load the original DataFrame from the Excel file
original_df = pd.read_excel(os.path.join(output_folder, 'Second_Transformed_Criteria_Tmin_Growing_season_'+scenario+'.xlsx'))

# Exclude the last column from the DataFrame
original_df = original_df.iloc[:, :-1]

# Calculate column-wise averages
RPI = original_df.mean()


sRPI = (RPI - RPI.min()) / (RPI.max() - RPI.min())

# Create DataFrames for RPI and sRPI
RPI_df = pd.DataFrame(data=RPI, columns=['RPI'])
sRPI_df = pd.DataFrame(data=sRPI, columns=['sRPI'])

# Add a new column for the order of sRPI
sRPI_df['Order'] = sRPI_df['sRPI'].rank()

# Merge DataFrames into a single DataFrame
combined_df = pd.concat([RPI_df, sRPI_df], axis=1)
# Sort the DataFrame based on the "Order" column in descending order
combined_df = combined_df.sort_values(by='Order', ascending=False)

# Export to the same Excel file with different columns
output_file = os.path.join(output_folder, 'RPI_sRPI_Tmin_Growing_season_'+scenario+'.xlsx')
combined_df.to_excel(output_file, index=True)
print("Column averages have been exported to RPI_"+scenario+".xlsx")
```

```python
%%script echo skipping
# @title Total sRPI (το τρέχω χωριστά)
import pandas as pd
import glob
scenario="rcp85"
# Step 1: Combine Excel files into one DataFrame
pattern = "*RPI_sRPI*"
matching_files = glob.glob(output_folder + "/" + pattern + "_"+scenario+".xlsx")
dfs = []

for file in matching_files:
    df = pd.read_excel(file)
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

# Step 2: Calculate the mean 'sRPI' for each unique value in 'Climate_model'
result_df = combined_df.groupby('Climate_model')['sRPI'].agg(['mean']).reset_index()

# Step 3: Rename the column to 'Average_sRPI'
result_df = result_df.rename(columns={'mean': 'Average_sRPI'})

# Step 4: Calculate the standardized_sRPI
min_value = result_df['Average_sRPI'].min()
max_value = result_df['Average_sRPI'].max()
result_df['standardized_sRPI'] = (result_df['Average_sRPI'] - min_value) / (max_value - min_value)

# Step 5: Add a column with the order of standardized_sRPI from smallest to largest
result_df['Order_sRPI'] = result_df['standardized_sRPI'].rank()

# Step 6: Sort the entire DataFrame based on the 'order_standardized_sRPI' column
result_df = result_df.sort_values(by='Order_sRPI',ascending=False)

# Step 5: Save the result DataFrame to a new Excel file
output_file = os.path.join(output_folder, "Ranking_results_"+scenario+".xlsx")
result_df.to_excel(output_file, index=False)

# Print the result
print(result_df)
```
