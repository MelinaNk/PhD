```python
%%script echo skipping
# @title Julian dates calculation
import datetime

file_name = os.path.join(combined_directory,Location+"_combined_file.txt")  # File name

 #Read the file
with open(file_name, 'r') as file:
    lines = file.readlines()

# Process each line
updated_lines = []
for line in lines:
    year= line[:2]
    date_str = line[:2] + "-" + line[2:4] + "-" + line[4:6]  # Extract year, month, and day from the line
    date = datetime.datetime.strptime(date_str, "%y-%m-%d")  # Convert date string to datetime object
    julian_date = date.timetuple().tm_yday  # Calculate Julian date
    updated_line = year+str(julian_date).zfill(3) + "\t" + line[7:]  # Create updated line with Julian date
    updated_lines.append(updated_line)

# Write updated lines to a new file
new_file_name = "updated_example.txt"
with open(new_file_name, 'w') as new_file:
    new_file.writelines(updated_lines)
```

```python
# @title Growing season averages
%%script echo skipping
import os
import pandas as pd


# Iterate through the files in the folder
for filename in os.listdir(combined_directory):
    if filename.endswith('combined_file.csv'):
        file_path = os.path.join(combined_directory, filename)
        df = pd.read_csv(file_path)

        # Convert Month and Year columns to numeric
        df['Month'] = pd.to_numeric(df['Month'], errors='coerce')
        df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
        df['Tmean']=pd.to_numeric(df['Tmean'], errors='coerce')
        df['Tmax']=pd.to_numeric(df['Tmax'], errors='coerce')
        df['Tmin']=pd.to_numeric(df['Tmin'], errors='coerce')
        df['Prec']=pd.to_numeric(df['Prec'], errors='coerce')



        # Filter the DataFrame for months 1, 2, and 12
filtered_df = df[df['Month'].isin([1, 2, 12])]

# Adjust the year for December (Month 12) to belong to the previous year
filtered_df.loc[filtered_df['Month'] == 12, 'Year'] += 1
# Add a column 'pr_no' representing the number of 'pr' values > 1
filtered_df['Prec_no'] = (filtered_df['Prec'] > 1).astype(int)


# Calculate the mean values grouped by Climate model and Year for months 1, 2, and 12
mean_values_by_year = filtered_df.groupby(['Year']).agg({
    'Tmax': 'mean',
    'Tmin': 'mean',
    'Tmean': 'mean',
    'Prec':  lambda x: x[x > 1].mean() if any(x > 1) else x.mean(),
    'Prec_no':'sum'# Apply correction for 'pr' values > 1
}).reset_index()

# Filter the DataFrame for months 3, 4, and 5
filtered_df_months_3_4_5 = df[df['Month'].isin([3, 4, 5])]
# Add a column 'pr_no' representing the number of 'pr' values > 1
filtered_df_months_3_4_5['Prec_no'] = (filtered_df_months_3_4_5['Prec'] > 1).astype(int)

# Calculate the mean values grouped by Climate model and Year for months 3, 4, and 5
mean_values_by_year_months_3_4_5 = filtered_df_months_3_4_5.groupby(['Year']).agg({
    'Tmax': 'mean',
    'Tmin': 'mean',
    'Tmean': 'mean',
    'Prec':  lambda x: x[x > 1].mean() if any(x > 1) else x.mean(),
    'Prec_no':'sum'# Apply correction for 'pr' values > 1
}).reset_index()

# Filter the DataFrame for months 6, 7
filtered_df_months_6_7 = df[df['Month'].isin([6, 7])]
# Add a column 'pr_no' representing the number of 'pr' values > 1
filtered_df_months_6_7['Prec_no'] = (filtered_df_months_6_7['Prec'] > 1).astype(int)

# Calculate the mean values grouped by Climate model and Year for months 6, 7
mean_values_by_year_months_6_7 = filtered_df_months_6_7.groupby(['Year']).agg({
    'Tmax': 'mean',
    'Tmin': 'mean',
    'Tmean': 'mean',
    'Prec':  lambda x: x[x > 1].mean() if any(x > 1) else x.mean(),
    'Prec_no':'sum'# Apply correction for 'pr' values > 1
}).reset_index()

# Filter the DataFrame for Growing season
filtered_df_gs = df[df['Month'].isin([1, 2, 3,4,5,6,7,12])]


# Adjust the year for December (Month 12) to belong to the previous year
filtered_df_gs.loc[filtered_df_gs['Month'] == 12, 'Year'] += 1
# Add a column 'pr_no' representing the number of 'pr' values > 1
filtered_df_gs['Prec_no'] = (filtered_df_gs['Prec'] > 1).astype(int)
# Calculate the mean values grouped by Climate model and Year for Growing season
mean_values_by_year_months_gr_seasons = filtered_df_gs.groupby(['Year']).agg({
    'Tmax': 'mean',
    'Tmin': 'mean',
    'Tmean': 'mean',
    'Prec':  lambda x: x[x > 1].mean() if any(x > 1) else x.mean(),
    'Prec_no':'sum'# Apply correction for 'pr' values > 1
}).reset_index()

# Merge the mean values for all month groups into a single DataFrame
final_mean_values = pd.merge(mean_values_by_year, mean_values_by_year_months_3_4_5, on=['Year'], suffixes=('', '_Spring'))
final_mean_values = pd.merge(final_mean_values, mean_values_by_year_months_6_7, on=['Year'], suffixes=('', '_Summer'))
final_mean_values = pd.merge(final_mean_values, mean_values_by_year_months_gr_seasons, on=['Year'], suffixes=('', '_Growing_season'))
final_mean_values['Location'] = Location2
final_mean_values['Elevation'] = elevation



# Specify the CSV file path
csv_file_path = os.path.join(meteo_directory, Location2+'_Growing_season_averages.csv')

# Save the DataFrame to CSV
final_mean_values.to_csv(csv_file_path, index=False)

print("Monthly averages saved to CSV:", csv_file_path)
```
